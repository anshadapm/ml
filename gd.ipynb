```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
x= np.random.rand(100)
c= np.random.rand(100)
y= []

for i in range(100):
    y_= 3*x[i] + c[i] 
    y.append(y_)
plt.scatter(x,y)

from sklearn.linear_model import LinearRegression
x=np.array(x).reshape((-1,1))
regr = LinearRegression().fit(x,y)
plt.plot
plt.plot(x,regr.predict(x),color='k')
print('slope',regr.coef_)
print('intercept',regr.intercept_)
        
```

```python
#sgd
plt.scatter(x,y)
m=random.random()*10
a=random.random()*10
alpha=0.2
j=0
for i in range(100):
    for k in range(100):
        j=0
        j+=1/100*pow(x[k]*m+a-y[k],2)
        dm=0
        da=0
        dm+=2*(m*x[k]+a-y[k])*x[k]
        da+=2*(m*x[k]+a-y[k])
        m=m-alpha*dm
        a=a-alpha*da   
plt.plot(x,x*m+a,'k')
plt.show()
print('slope',m)
print('intercept',a)
```

```python
#batch gd
plt.scatter(x,y)
m=random.random()*10
a=random.random()*10
alpha=0.3
j=0
for i in range(100):
    for k in range(100):
        j=0
        j+=1/100*pow(x[k]*m+a-y[k],2)
    dm=0
    da=0
    for l in range(100): 
        dm+=2/100*(m*x[l]+a-y[l])*x[l]
        da+=2/100*(m*x[l]+a-y[l])
        m=m-alpha*dm
        a=a-alpha*da   
plt.plot(x,x*m+a,'k')
plt.show()
print('slope',m)
print('intercept',a)
```

```python
#mini batch gd
x= np.random.rand(100)
c= np.random.rand(100)
y= np.zeros(100)

for i in range(100):
    y[i]= 3*x[i] + c[i] 
plt.scatter(x,y)

x.resize(10,10)
y.resize(10,10)

m=random.random()*10
a=random.random()*10
alpha=0.1
j=0
for i in range(100):
    for k in range(10):
        j=0
        for n in range(10):
            j+=1/100*pow(x[k][n]*m+a-y[k][n],2)
        dm=0
        da=0
        for l in range(10): 
            dm+=2/10*(m*x[k][l]+a-y[k][l])*x[k][l]
            da+=2/10*(m*x[k][l]+a-y[k][l])
        m=m-alpha*dm
        a=a-alpha*da   
plt.plot(x,x*m+a,'k')
plt.show()
print('slope',m)
print('intercept',a)
```
