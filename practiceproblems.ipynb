Sampling from a distribution. Use the Python libraries numpy and matplotlib

(a)

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
mean=[0, 0]
cov=[[1,0],[0,1]]
(x1,x2)=np.random.multivariate_normal(mean,cov,100).T
plt.plot(x1,x2,'o')
plt.show()
```
